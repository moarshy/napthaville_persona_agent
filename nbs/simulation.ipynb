### Simulation

```python
isabella = Persona(
    "Isabella Rodriguez", 
    memory_deployment=memory_deployment,
    maze_deployment=env_deployment
)

maria = Persona(
    "Maria Lopez", 
    memory_deployment=memory_deployment,
    maze_deployment=env_deployment
)

klaus = Persona(
    "Klaus Mueller", 
    memory_deployment=memory_deployment,
    maze_deployment=env_deployment
)

await isabella.load_memory()
await maria.load_memory()
await klaus.load_memory()
```

```python
current_positions = {
  "Isabella Rodriguez": [73, 14],
  "Klaus Mueller": [127, 46],
  "Maria Lopez": [123, 57]
}

new_positions = {
    "Isabella Rodriguez": [73, 14],
    "Klaus Mueller": [127, 46],
    "Maria Lopez": [123, 57]
}
```

```python
personas = {
    "Isabella Rodriguez": isabella,
    "Klaus Mueller": klaus,
    "Maria Lopez": maria
}
persona_names = list(personas.keys())
step = 0
curr_time = datetime.now()
seconds_per_step = 60
```

```python
game_obj_cleanup = dict()
# TODO: need to do this but we skip for now: self.maze.turn_event_from_tile_idle(key, val)

for persona_name, persona in personas.items(): 
    curr_tile = [current_positions[persona_name][0], current_positions[persona_name][1]]
    new_tile = [new_positions[persona_name][0], new_positions[persona_name][1]]

    # remove subject from events from current tile
    maze_call = {
        "inputs": {
            "function_name": "remove_subject_events_from_tile",
            "function_input_data": {
                "subject": persona_name,
                "tile": {"x": curr_tile[0], "y": curr_tile[1]}
            }
        },
        "deployment": env_deployment
    }
    result = await env_run(maze_call)
    print(f"removed subject events from tile {curr_tile}")

    # add event from tile 
    maze_call = {
        "inputs": {
            "function_name": "add_event_from_tile",
            "function_input_data": {
                "tile": {"x": new_tile[0], "y": new_tile[1]},
                "curr_event": persona.scratch.get_curr_event_and_desc()
            }
        },
        "deployment": env_deployment
    }
    result = await env_run(maze_call)
    print(f"added event from tile {new_tile}")

    if not persona.scratch.planned_path:
        game_obj_cleanup[persona.scratch.get_curr_obj_event_and_desc()] = new_tile

        maze_call = {
            "inputs": {
                "function_name": "add_event_from_tile",
                "function_input_data": {
                    "tile": {"x": new_tile[0], "y": new_tile[1]},
                    "curr_event": persona.scratch.get_curr_obj_event_and_desc()
                }
            },
            "deployment": env_deployment
        }
        result = await env_run(maze_call)
        print(f"added event from tile {new_tile}")
        
        blank = (persona.scratch.get_curr_obj_event_and_desc()[0], None, None, None)
        maze_call = {
            "inputs": {
                "function_name": "remove_event_from_tile",
                "function_input_data": {
                    "tile": {"x": new_tile[0], "y": new_tile[1]},
                    "curr_event": blank
                }
            },
            "deployment": env_deployment
        }
        result = await env_run(maze_call)
        print(f"removed event from tile {new_tile}")
    
    movements = {
        "persona" : {},
        "meta": {}
    }

    for persona_name, persona in personas.items():
        next_tile, pronunciatio, description = await persona.move(
            persona_names,
            curr_tile=curr_tile,
            curr_time=curr_time
        )

        movements["persona"][persona_name] = {}
        movements["persona"][persona_name]["movement"] = next_tile
        movements["persona"][persona_name]["pronunciatio"] = pronunciatio
        movements["persona"][persona_name]["description"] = description
        movements["persona"][persona_name]["chat"] = (persona.scratch.chat)
        movements["meta"]["curr_time"] = (curr_time.strftime("%B %d, %Y, %H:%M:%S"))

    # TODO:SAVE MOVEMENTS TO DB
    
    step += 1
    curr_time += datetime.timedelta(seconds=seconds_per_step)
```
